```python
co2 = dataset['co2']
# print(co2.min())
# print(co2.max())
# print(dataset.shape)
bins = [20, 60, 100, 140, 180, 220, 260, 300, 340, 380, 420, 460, 500, 540, 580, 620, 660, 700]
plt.style.use('fivethirtyeight')
plt.hist(co2, bins=bins, width=35, color="navy")
plt.title('Emission of Carbone Dioxide emitted by Vehicles')
plt.xlabel('Co2 Emissions')
plt.ylabel('Number of Vehicles')
plt.show()


```

This Histogram shows carbon dioxide emmitted by diffrent vehicles

```python
feScore = dataset['feScore']
# print(feScore.min())
# print(feScore.max())
plt.style.use('fivethirtyeight')
plt.hist(feScore, bins=10, width=0.8, color="navy")
plt.title('Fuel Economy Score of Vehicles')
plt.xlabel('feScore')
plt.ylabel('Number of Vehicles')
plt.show()
```

This Histogram shows feScore of diffrent vehicles, where a vehicles with high feScore indicates a better fuel economy, meaning it can travel travel farther on less fuel.

```python
cylinders = dataset['cylinders']
# print(cylinders.min())
# print(cylinders.max())
plt.style.use('fivethirtyeight')
bins = [2, 4, 6, 8, 10, 12]
plt.hist(cylinders, bins=bins, width=1.8, color="navy")
plt.title('Number of Cylinders a Vehicle has')
plt.xlabel('Cylinders')
plt.ylabel('Number of Vehicles')
plt.show()
```

This Histogram show how many vehicles has different cylinders

```python
numeric_data = dataset.select_dtypes(include=['number'])
# print(numeric_data)
numdata = numeric_data.corr()
plt.figure(figsize=(9,7))
sns.heatmap(numdata, annot=True, cmap='YlGnBu')
plt.title('Vehicles Correlation Heat Map')
plt.show()
```

**Heatmap Interpretation**

This Heatmap shows some of the attributes of different vehicles using different colors starting from least heated with light yellow to the heavy heated with dark blue.  

**Columns that can be deleted**
